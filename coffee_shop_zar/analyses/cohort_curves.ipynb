```python
from google.cloud import bigquery
import pandas as pd
from plotnine import *
```

```python
PROJECT_ID = "aec-students"
client = bigquery.Client(PROJECT_ID)
```

```python
df = client.query("select * from dbt_lauren.ltv_cohorts").to_dataframe()
```

```python
df.head()
```

```python
(
    ggplot(df, aes(x="week", y="average_ltv", color="cohort_week"))
    + geom_line()
    + theme_bw()
    + theme(figure_size=(10, 8))
)
```
